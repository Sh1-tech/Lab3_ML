```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print(numeric_cols)


df[numeric_cols].describe()
```

```python
print("Variance:\n", df[numeric_cols].var())
print("\nStandard Deviation:\n", df[numeric_cols].std())
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['User_Rating'], bins=20)
plt.title("User_Rating")
plt.show()
```

```python
print("Connor's name", df.columns.tolist())
print(df.head())
df['Tool_Name'] = df['Tool_Name']



traffic_by_tool = df.groupby('Tool_Name')['Monthly_Traffic_Est'].sum()

plt.figure(figsize=(10,5))
traffic_by_tool.plot(kind='bar') 
plt.title("Total Monthly Traffic by Tool")
plt.ylabel("Traffic")
plt.xlabel("Tool Name")
plt.xticks(rotation=60)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['User_Rating', 'Monthly_Traffic_Est']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Starting_Price_USD', y='User_Rating', data=df)
plt.title("Boxes Shipped vs Revenue")
plt.show()
```

```python
corr_matrix = df[numeric_cols].corr()


plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('corr_matrix')
plt.show()
```
